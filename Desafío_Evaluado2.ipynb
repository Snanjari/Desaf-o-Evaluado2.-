Genere funciones para calcular la media y varianza de un vector. Debe cumplir con
los siguientes requisitos:
    
○ Ambas funciones deben ingresar un argumento x correspondiente al vector.

○ Las funciones deben contener docstrings con la documentación asociada a
la variable y retornar un valor utilizando return.

○ La función de la varianza se debe llamar a la función de la media.

● Utilice las funciones para reportar la información sobre goles_favor, goles_contra
y puntos.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('worldcup2014.csv')
df
```

```python
def mean(x):
    """
    Calcular la media de un vector x.
    
    Parameters
    ----------
    x : list or numpy array
        Vector de datos.
        
    Returns
    -------
    media : float
        Media del vector x.
    """
    return sum(x) / len(x)
```

```python
def variance(x):
    """
    Calcula la varianza de un vector x.
    
    Parameters
    ----------
    x : list or numpy array
        Vector de datos.
        
    Returns
    -------
    varianza : float
        Varianza del vector x.
    """
    media = mean(x)
    return sum((xi - media)**2 for xi in x) / len(x)
```

```python
"""
La función variance toma como argumento un vector x y utiliza la función mean para calcular la media de x. 
Luego, utiliza la fórmula sum((xi - media)**2 for xi in x) / len(x) para calcular la varianza de x.
Podemos utilizar estas funciones para calcular la media y la varianza de los vectores goles_favor, goles_contra y puntos 
de la siguiente manera:
"""

goles_favor = df['goles_favor']
goles_contra = df['goles_contra']
puntos = df['puntos']

print("Media goles_favor: ", mean(goles_favor))
print("Varianza goles_favor: ", variance(goles_favor))
print(" ")
print("Media goles_contra: ", mean(goles_contra))
print("Varianza goles_contra: ", variance(goles_contra))
print(" ")
print("Media puntos: ", mean(puntos))
print("Varianza puntos: ", variance(puntos))
```

2) Desafío 2:

Utilizando el método groupby de la clase DataFrame, en conjunto con la función .agg de
pandas, calcule la media, la varianza y desviación estándar de la cantidad de goles a favor,
en contra y de la cantidad de puntos por continente.

● ¿En qué continente se observa una mayor cantidad de goles a favor?

● ¿En qué continente se observa una mayor cantidad de goles en contra?

● ¿En qué continente se observa una mayor cantidad de puntos en promedio?


```python
# Calculamos la media, la varianza y la desviación estándar por continente
result = df.groupby("continent").agg({
    'goles_favor': ["mean", "var", "std"],
    'goles_contra': ["mean", "var", "std"],
    'puntos': ["mean", "var", "std"]
})

print(result)
```

```python
max_continent = result['goles_favor']['mean'].idxmax()
print(f'El continente con la mayor cantidad de goles a favor es {max_continent}')
```

```python
less_continent = result['goles_contra']['mean'].idxmax()
print(f'El continente con la mayor cantidad de goles en contra es {less_continent}')
```

```python
mean_points = result['puntos']['mean'].idxmax()
print(f'El continente con la mayor cantidad de puntos en promedio es {mean_points}')
```

4) Desafío 4: Función simuladora: 

Genere una función llamada simulate_pets_prob que tome como argumento un
número finito de simulaciones a generar.

● La función debe simular dos situaciones young_pet y old_pet, y contar la
ocurrencia de los siguientes casos:

○ De las dos mascotas simuladas (young y old), contar las ocasiones donde
por lo menos una de las mascotas sea un perro.

○ De las dos mascotas simuladas, contar las ocasiones donde old_pet sea un
perro.

○ De las dos mascotas simuladas, contar las ocasiones donde los dos sean
perros.

● El método debe tener una semilla pseudoaleatoria de 1.

● El output de la función debe entregarse en términos de Probabilidad.

● De los tres escenarios, ¿Cuál es el menos probable? ¿Cuál es el más probable? ¿Por
qué?


```python
def simulate_pets_prob(n_simulations):
    np.random.seed(1)
    
    young_pets = generate_pet(n_simulations)
    old_pets = generate_pet(n_simulations)
    
    at_least_one_dog = np.sum(np.logical_or(young_pets == 'perro', old_pets == 'perro'))
    old_pet_dog = np.sum(old_pets == 'perro')
    both_dogs = np.sum(np.logical_and(young_pets == 'perro', old_pets == 'perro'))
    
    prob_at_least_one_dog = at_least_one_dog / n_simulations
    prob_old_pet_dog = old_pet_dog / n_simulations
    prob_both_dogs = both_dogs / n_simulations
    
    return prob_at_least_one_dog, prob_old_pet_dog, prob_both_dogs
```
